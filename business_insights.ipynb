```python
# Box Plots for R, F, M
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.boxplot(y=rfm_data["Recency"])
plt.title("Recency")
plt.ylabel("Days")

plt.subplot(1, 3, 2)
sns.boxplot(y=rfm_data["Frequency"])
plt.title("Frequency")
plt.ylabel("Number of Purchases")

plt.subplot(1, 3, 3)
sns.boxplot(y=rfm_data["Monetary"])
plt.title("Monetary")
plt.ylabel("Amount Spent")
```

```python
# Histogram Plots for R, F, M
plt.figure(figsize=(15, 5)) 
plt.subplot(1, 3, 1)
sns.histplot(rfm_data["Recency"], bins=30, kde=True)
plt.title("Recency Distribution")
plt.xlabel("Days Since Last Purchase")
plt.ylabel("Frequency")

plt.subplot(1, 3, 2)
sns.histplot(rfm_data["Frequency"], bins=30, kde=True)
plt.title("Frequency Distribution")
plt.xlabel("Number of Purchases")
plt.ylabel("Frequency")

plt.subplot(1, 3, 3)
sns.histplot(rfm_data["Monetary"], bins=30, kde=True)
plt.title("Monetary Distribution")
plt.xlabel("Amount Spent")
plt.ylabel("Frequency")
```
